```python
import pandas as pd
```

```python
dict1={'names':['ajay','sanjay','raheem','sudha','loha'],'course':['python','python','java','R','R'],'marks':[7,8,2,1,5]}
dict2={'names':['basha','sam','suji'],'course':['java','DA','AI'],'marks':[4,7,8],'location':['mumbai','calcutta','orissa']}

dict3= {'names':['sanjay','raheem','sudha','raja'],'groups':[1,2,3,4]}
```

### Concat

```python
df1= pd.DataFrame(dict1)
df2= pd.DataFrame(dict2)
df3= pd.DataFrame(dict3)
```

```python
df2
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

## Merge

```python
df1
```

```python
## inner Join
pd.merge(df1,df3,on='names',how='inner')
```

```python
## left_join
pd.merge(df1,df3,on='names',how='left')
```

```python
## Right_join
pd.merge(df1,df3,on='names',how='right')
```

```python
## outer_join
pd.merge(df1,df3,on='names',how='outer')
```

## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Exp=[1,2,3,4,5]
Salary=[1000,2000,3003,4500,5600]
```

```python
plt.figure(figsize=(8,5),facecolor='lightgreen')
plt.plot(Exp,Salary,marker='o',markersize=10,markerfacecolor='red',color='seagreen',linewidth=2,linestyle=':')
plt.xlabel('Experience',size=15,color='red')
plt.ylabel('Salary',size=15,color='red')
plt.title('Exp Vs Salary',size=20,color='red')
plt.grid()
```

```python
fig,axs= plt.subplots(2,3,figsize=(14,8))
plt.suptitle('Peformance_Measures',size=25,color='green',fontweight='bold',y=(1))
axs[0,1].plot(Exp,Salary)
axs[0,1].set_xlabel('Exp',color='red',size=10)
axs[0,1].set_ylabel('Salary',color='red',size=10)
axs[0,1].set_title('Exp vs Salary',color='red')
axs[1,0].bar(Exp,Salary)
axs[1,0].set_xlabel('Exp',color='red',size=10)
axs[1,0].set_ylabel('Salary',color='red',size=10)
axs[1,0].set_title('Exp vs Salary',color='red')
axs[1,2].scatter(Exp,Salary)
axs[1,2].set_xlabel('Exp',color='red',size=10)
axs[1,2].set_ylabel('Salary',color='red',size=10)
axs[1,2].set_title('Exp vs Salary',color='red')
plt.tight_layout()
```

```python
a=plt.bar(Exp,Salary,width=0.9,edgecolor='black')
plt.xlabel('Experience',size=15,color='red')
plt.ylabel('Salary',size=15,color='red')
plt.title('Exp Vs Salary',size=20,color='red')
plt.bar_label(a);
```
