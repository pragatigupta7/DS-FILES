```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('DB_dataset.csv')
```

```python
df.head()
```

```python
std_sca=StandardScaler()
```

```python
df1=pd.DataFrame(std_sca.fit_transform(df),columns=df.columns)
df1
```

```python
iso=IsolationForest(n_estimators=100,contamination=0.07)
```

```python
data=iso.fit(df1)
data
```

```python
pred=data.predict(df1)
pred
```

```python
index=np.where(pred<0)
index
```

```python
x=df1.values
x
```

```python
plt.scatter(x[:,0],x[:,1])
plt.scatter(x[index,0],x[index,1],color='red')
plt.xlabel('Weight')
plt.ylabel('Height')
```
