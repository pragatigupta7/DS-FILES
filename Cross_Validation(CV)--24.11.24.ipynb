```python
from sklearn.datasets import load_iris
from sklearn.model_selection import LeaveOneOut,KFold,cross_val_score,StratifiedKFold
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
my_data= load_iris()
```

```python
features= pd.DataFrame(my_data.data,columns=my_data.feature_names)
features
```

```python
target=pd.DataFrame(my_data.target,columns=['Target'])
target
```

```python
leave= LeaveOneOut()
```

```python
leave.split(features,target)
```

```python
for train_split ,test_split in leave.split(features,target):
    print(train_split,test_split)
```

```python
kfold=KFold(n_splits=5,shuffle=True)
```

```python
for train_split,test_split in kfold.split(features,target):
    print(train_split,test_split)
```

```python
str_kold= StratifiedKFold(n_splits=5,shuffle=True)
```

```python
for train_split,test_split in str_kold.split(features,target):
    print(train_split,test_split)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
cross_val_score(DecisionTreeClassifier(),features,target,scoring='accuracy',cv=10).mean()
```
